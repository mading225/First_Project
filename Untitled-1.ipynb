DATA CLEANING AND ANALYSIS

---



```python
# dropping irrelevent columns 
columns_to_drop = [
    'Latitude',
    'Longitude',
    'FAR.Description',
    'Schedule',
    'Air.carrier',
    'Airport.Name',
    'Airport.Code',
    'Registration.Number',
    'Event.Id',
    
    'Publication.Date',
    'Investigation.Type',
    'Accident.Number',
    
    
    

]
df.drop(columns=columns_to_drop, inplace=True)
df.shape
```

```python
# Feeling in the missing values of the relevant columns 
df['Total.Uninjured'] = df['Total.Uninjured'].fillna(df['Total.Uninjured'].mean())
df['Total.Fatal.Injuries'] = df['Total.Fatal.Injuries'].fillna(df['Total.Fatal.Injuries'].mean())
df['Total.Serious.Injuries'] = df['Total.Serious.Injuries'].fillna(df['Total.Serious.Injuries'].mean()) 
df['Total.Minor.Injuries'] = df['Total.Minor.Injuries'].fillna(df['Total.Minor.Injuries'].mean())


df['Event.Date'] = df['Event.Date'].fillna('unknown')
                                                        

df['Injury.Severity'] = df['Injury.Severity'].fillna(0)
df['Aircraft.Category'] = df['Aircraft.Category'].fillna('unspecified')
df['Amateur.Built'] = df['Amateur.Built'].fillna(0)
df['Number.of.Engines'] = df['Number.of.Engines'].fillna(df['Number.of.Engines'].mean())
df['Engine.Type'] = df['Engine.Type'].fillna(0)
df['Purpose.of.flight'] = df['Purpose.of.flight'].fillna(0)
df['Weather.Condition'] = df['Weather.Condition'].fillna(0) 

df['Model'] = df['Model'].fillna(0)
df['Location'] = df['Location'].fillna(0)
df['Make'] = df['Make'].fillna(0)

df['Country'] = df['Country'].fillna(0)
df['Report.Status'] = df['Report.Status'].fillna(0)
df['Aircraft.damage'] = df['Aircraft.damage'].fillna(0)
df['Make'] = df['Make'].fillna(0)
print(df)
```

```python
# checking if the missing values have been cleaned 
df.isna().sum()
```

```python
# Dropping the Broad.phase.of.flight column 
df = df.dropna(subset = ['Broad.phase.of.flight'])
```

```python
# These are the top 20 aircraft make and model total fatal injuries 
# The code below is showing the number of fatal injuries that occured during the aircraft crash per make and model of the aircraft
make_model_fatalities = df.groupby(['Make', 'Model'])['Total.Fatal.Injuries'].sum().reset_index()
make_model_fatalities = make_model_fatalities.sort_values(by='Total.Fatal.Injuries', ascending=False).head(20)

print("Top 20 Make-Model Combinations by Total Fatal Injuries:")
print(make_model_fatalities)
```

```python
# Visualizing the total fatal injusries using the make and model of the aircraft
#Cessna-152 had the most fata injuries 
make_model_fatalities['Make_Model'] = make_model_fatalities['Make'] +  make_model_fatalities['Model']

plt.figure(figsize=(14, 8))
sns.barplot(x='Make_Model', y='Total.Fatal.Injuries', data=make_model_fatalities, palette='viridis')
plt.xticks(rotation=90, ha='center')
plt.title('Top 20 Make-Model Combinations by Total Fatal Injuries')
plt.xlabel('Aircraft Make - Model')
plt.ylabel('Total Fatal Injuries')
plt.tight_layout()
plt.show()
```
