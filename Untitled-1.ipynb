DATA CLEANING AND ANALYSIS

---



```python
# dropping irrelevent columns 
columns_to_drop = [
    'Latitude',
    'Longitude',
    'FAR.Description',
    'Schedule',
    'Air.carrier',
    'Airport.Name',
    'Airport.Code',
    'Registration.Number',
    'Event.Id',
    
    'Publication.Date',
    'Investigation.Type',
    'Accident.Number'
    
    

]
df.drop(columns=columns_to_drop, inplace=True)
df.shape
```

```python
# Feeling in the missing values of the relevant columns 
df['Total.Uninjured'] = df['Total.Uninjured'].fillna(df['Total.Uninjured'].mean())
df['Total.Fatal.Injuries'] = df['Total.Fatal.Injuries'].fillna(df['Total.Fatal.Injuries'].mean())
df['Total.Serious.Injuries'] = df['Total.Serious.Injuries'].fillna(df['Total.Serious.Injuries'].mean()) 
df['Total.Minor.Injuries'] = df['Total.Minor.Injuries'].fillna(df['Total.Minor.Injuries'].mean())


df['Event.Date'] = df['Event.Date'].fillna('unknown')
                                                        

df['Injury.Severity'] = df['Injury.Severity'].fillna(0)
df['Aircraft.Category'] = df['Aircraft.Category'].fillna('unspecified')
df['Amateur.Built'] = df['Amateur.Built'].fillna(0)
df['Number.of.Engines'] = df['Number.of.Engines'].fillna(df['Number.of.Engines'].mean())
df['Engine.Type'] = df['Engine.Type'].fillna(0)
df['Purpose.of.flight'] = df['Purpose.of.flight'].fillna(0)
df['Weather.Condition'] = df['Weather.Condition'].fillna(0) 

df['Model'] = df['Model'].fillna(0)
df['Location'] = df['Location'].fillna(0)
df['Make'] = df['Make'].fillna(0)

df['Country'] = df['Country'].fillna(0)
df['Report.Status'] = df['Report.Status'].fillna(0)
df['Aircraft.damage'] = df['Aircraft.damage'].fillna(0)
df['Make'] = df['Make'].fillna(0)
print(df)
```
