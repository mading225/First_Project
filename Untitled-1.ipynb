## Calculating Incident and Injury Metrics per Aircraft Make

### 
This is done by Grouping the DataFrame by 'Make' to calculate the total number of incidents, total fatal injuries, total serious injuries, and total minor injuries for each aircraft make. Also, counting the incidents that resulted in 'Destroyed' or 'Substantial' aircraft damage for each make.


**Explanation**:
This is done by  calculating the total number of incidents, various injury totals, and the count of 'Destroyed' or 'Substantial' aircraft damage per aircraft make.This is by grouping the DataFrame by 'Make' and applying aggregation functions.



```python
make_metrics = df.groupby('Make').agg(
    total_incidents=('Make', 'size'), # The size element is counting how many times each manufacturer is appearing in the data set 
    total_fatal_injuries=('Total.Fatal.Injuries', 'sum'),
    total_serious_injuries=('Total.Serious.Injuries', 'sum'),
    total_minor_injuries=('Total.Minor.Injuries', 'sum'),
    incidents_severe_damage=(
        'Aircraft.damage',
        lambda x: ((x == 'Destroyed') | (x == 'Substantial')).sum()
    )
).reset_index()

print(make_metrics.head())
```


The incident and injury metrics have been calculated for each aircraft make, this is done by  computing the fatal injury rate and the severe damage rate for each make. This will involve dividing the total fatal injuries and severe damage incidents by the total number of incidents for each make.



```python
make_metrics['fatal_injury_rate'] = make_metrics['total_fatal_injuries'] / make_metrics['total_incidents']
make_metrics['severe_damage_rate'] = make_metrics['incidents_severe_damage'] / make_metrics['total_incidents']

print(make_metrics.head())
```

**Explanation**:
Identify the lowest risk aircraft makes, this is done by  first filter out makes with very low number of incidents to ensure the rates are statistically significant. Then,ing calculating  an overall risk score using both 'fatal_injury_rate' and 'severe_damage_rate' and sort the DataFrame to find the top 10 lowest risk makes.



```python
min_incidents = 10
filtered_make_metrics = make_metrics[make_metrics['total_incidents'] >= min_incidents]

# Creating a composite risk score (e.g., average of normalized rates)
# Normalizing rates to ensure they have similar scale if needed
filtered_make_metrics['overall_risk_score'] = (filtered_make_metrics['fatal_injury_rate'] + filtered_make_metrics['severe_damage_rate']) / 2

# Sorting by overall risk score to find the lowest risk makes
lowest_risk_makes = filtered_make_metrics.sort_values(by='overall_risk_score').head(10)

print("Top 10 Lowest Risk Aircraft Makes based on composite risk score:")
print(lowest_risk_makes[['Make', 'total_incidents', 'fatal_injury_rate', 'severe_damage_rate', 'overall_risk_score']])
```

**Reasoning**:
To compare the risk by purpose of flight, the df_low_risk_makes' DataFrame should include only 'Personal' and 'Business' flights, as these are the primary categories of interest for the project objectives. This will prepare the data for further aggregation and risk calculation based on these specific purposes of flight.



```python
# 1. Getting  the list of the top 10 safest makes from the summary table
safe_makes = lowest_risk_makes['Make'].unique()

df_low_risk_data = df[df['Make'].isin(safe_makes)]

# 2 filtering for Business and Personal flight purposes
df_low_risk_purpose = df_low_risk_data[df_low_risk_data['Purpose.of.flight'].isin(['Business', 'Personal'])]

# 3. Checking results
print(df_low_risk_purpose['Purpose.of.flight'].value_counts())
```

### Observation on Purpose of Flight

Among the 'lowest_risk_makes' identified, 'Personal' flights have a lower risk compared to the ('commercial' aka 'Business') in the 'Purpose.of.flight' column within the filtered 'df_low_risk_purpose' DataFrame.

```python
# Calculating incident and injury matrices for these flights within low risk aircraft make
purpose_flight_metrics = df_low_risk_purpose.groupby(['Make', 'Purpose.of.flight']).agg(
    total_incidents=('Make', 'size'),
    total_fatal_injuries=('Total.Fatal.Injuries', 'sum'),
    total_serious_injuries=('Total.Serious.Injuries', 'sum'),
    total_minor_injuries=('Total.Minor.Injuries', 'sum'),
    incidents_severe_damage=(
        'Aircraft.damage',
        lambda x: ((x == 'Destroyed') | (x == 'Substantial')).sum()
    )
).reset_index()

purpose_flight_metrics['fatal_injury_rate'] = purpose_flight_metrics['total_fatal_injuries'] / purpose_flight_metrics['total_incidents']
purpose_flight_metrics['severe_damage_rate'] = purpose_flight_metrics['incidents_severe_damage'] / purpose_flight_metrics['total_incidents']

print("Risk Metrics by Purpose of Flight for Low-Risk Aircraft Makes:")
print(purpose_flight_metrics.sort_values(by='fatal_injury_rate').head(10))
```


Visualizing the overall risk of the top 10 lowest risk aircraft makes, using a  bar plot showing their 'overall_risk_score'. This will provide a clear visual representation of their risk levels.



```python
plt.figure(figsize=(12, 7))
sns.barplot(x='Make', y='overall_risk_score', data=lowest_risk_makes.sort_values(by='overall_risk_score', ascending=True))
plt.xticks(rotation=45, ha='right')
plt.title('Top 10 Lowest Risk Aircraft Makes by Overall Risk Score')
plt.xlabel('Aircraft Make')
plt.ylabel('Overall Risk Score')
plt.tight_layout()
plt.show()
```

## Calculati Incident and Injury Metrics per Aircraft Make




**Reasoning**:
To calculate the incident and injury metrics for each aircraft make, group the DataFrame `df` by the 'Make' column and apply aggregation functions to count incidents, sum injuries, and count severe damage incidents.Then use a lambda function to calculate the 'incidents_severe_damage' correctly.



```python
make_metrics = df.groupby('Make').agg(
    total_incidents=('Make', 'size'), # The size element is counting how many times each manufacturer is appearing in the data set
    total_fatal_injuries=('Total.Fatal.Injuries', 'sum'),
    total_serious_injuries=('Total.Serious.Injuries', 'sum'),
    total_minor_injuries=('Total.Minor.Injuries', 'sum'),
    incidents_severe_damage=(
        'Aircraft.damage',
        lambda x: ((x == 'Destroyed') | (x == 'Substantial')).sum()
    )
).reset_index()

print(make_metrics.head())
```

## Calculating Risk Rates per Aircraft Make




**Reasoning**:
To calculate the risk rates per aircraft make, the make_marices is  filtered  for makes with more than 50 incidents, then computed to  the 'fatal_injury_rate' and 'severe_damage_rate' for these filtered makes, and finally display the head of the resulting DataFrame.



```python
min_incidents_threshold = 50
filtered_make_metrics = make_metrics[make_metrics['total_incidents'] > min_incidents_threshold].copy()

filtered_make_metrics['fatal_injury_rate'] = filtered_make_metrics['total_fatal_injuries'] / filtered_make_metrics['total_incidents']
filtered_make_metrics['severe_damage_rate'] = filtered_make_metrics['incidents_severe_damage'] / filtered_make_metrics['total_incidents']

print(filtered_make_metrics.head())
```
